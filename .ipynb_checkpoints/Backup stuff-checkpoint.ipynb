```python
# Check the ratio between cancelled and uncancelled bookings
not_cancelled = df['IsCanceled'].value_counts()[0]
cancelled = df['IsCanceled'].value_counts()[1]
total = not_cancelled + cancelled
print(f'Uncancelled Bookings: {(not_cancelled/(total)) * 100:.3f}%\nCancelled Bookings: {(cancelled/(total)) * 100:.3f}%')
```

```python
# this is probably not needed for this project
mean = []
std = []
cols = []

stats = []
for col in df.columns:
    if is_numeric_dtype(df[col]):
        temp = []
    
        temp.append(col)
        temp.append(df[col].mean())
        temp.append(df[col].std())
        
        stats.append(temp)

stats_df = pd.DataFrame(data=stats)
stats_df.columns = ['', 'Mean', 'Std. Dev.']
stats_df
```

```python
# Children and babies
plt.rcParams["figure.figsize"] = (20,10)

no_children = df['Children'].value_counts()[0]
with_children = sum(x for x in df['Children'].value_counts()[1:])
total = not_cancelled + cancelled
print(f'Guests with 1 or more children: {(with_children/(total)) * 100:.3f}%\nGuests without children: {(no_children/(total)) * 100:.3f}%')
print() 

cancelled_without_children = 0
children_array = np.array(df['Children'].to_list()).astype(np.float)
for i in range(0, len(children_array)):
    if children_array[i] == 0 and df.iloc[i, 0] == 1:
        cancelled_without_children += 1
print(f'Cancelled bookings for guests with no children: {(cancelled_without_children/no_children) * 100:.3f}%')

cancelled_with_children = 0
for i in range(0, len(children_array)):
    if children_array[i] != 0 and df.iloc[i, 0] == 1:
        cancelled_with_children += 1
print(f'Cancelled bookings for guests with 1 or more children: {(cancelled_with_children/with_children) * 100:.3f}%')
print()


print("The pie chart below represents the distribution between the guests without children and ones with children.\nThe inner circle represents the amount of cancelled vs. uncancelled bookings for both of the categories")
labels = ["Without children", "With 1 or more children"]
fig, ax = plt.subplots()
size = 0.33
vals = np.array([[cancelled_without_children, no_children - cancelled_without_children], [cancelled_with_children, with_children - cancelled_with_children]])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(2)*4)
inner_colors = cmap([1, 2, 5, 6, 9, 10])

ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'), labels=labels, autopct='%1.1f%%')

ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title='Booking status based on children or no children')


fig.show()
```

## Handle Missing values

```python
# Remove Company feature, since there are too many missing values in this feature
# maybe only delete company from the training dataset
if "Company" in df:
    print("Deleted Company")
    df = df.drop(["Company"], axis=1)
```

```python
# TESTING STUFF HERE 
def splitDateString(value):
    splitted = value.split('-')
    return splitted

if 'ReservationStatusDate' in df.columns:
    # Split date string 
    df['ReservationStatusDate'] = df['ReservationStatusDate'].transform(splitDateString)
    
    # Store year, month, and day as separate features
    df['ReservationStatusYear'] = df['ReservationStatusDate'].str[0]
    df['ReservationStatusMonth'] = df['ReservationStatusDate'].str[1]
    df['ReservationStatusDay'] = df['ReservationStatusDate'].str[2]
    print('Transformed')
    
    # Drop original column which will not be used
    df = df.drop(['ReservationStatusDate'], axis=1)
    print('Value dropped')
else:
    print('Already transformed and dropped column')
    
df.ReservationStatusYear = df.ReservationStatusYear.astype(int)
df.dtypes
```

```python
# missing = df.index[df.Country.isnull().any(axis=1)]
# missing.shape
# missing = df.index[df.Country.isnull().any(axis=1)]
missing_country = df[df.Country.isnull()]
missing_agent = df[df.Agent.isnull()]
non_missing_values = df.dropna()

non_missing_values
```

```python
# using random forest to determine missing values
x_country = non_missing_values.drop(['Country'], axis=1)
y_country = non_missing_values['Country']
x_agent = non_missing_values.drop(['Agent'], axis=1)
y_agent = non_missing_values['Agent']

country_encoder = ce.OrdinalEncoder(cols=['ArrivalDateMonth', 'DepositType', 'CustomerType', 
                                  'ReservationStatus', 'ReservationStatusDate', 'Meal', 
                                  'MarketSegment', 'DistributionChannel', 
                                  'ReservedRoomType', 'ReservationStatus', 'AssignedRoomType', 
                                  'ReservedRoomType'])
agent_encoder = ce.OrdinalEncoder(cols=['ArrivalDateMonth', 'DepositType', 'CustomerType', 
                                  'ReservationStatus', 'ReservationStatusDate', 'Meal', 
                                  'Country', 'MarketSegment', 'DistributionChannel', 
                                  'ReservedRoomType', 'ReservationStatus', 'AssignedRoomType', 
                                  'ReservedRoomType'])


x_country = country_encoder.fit_transform(x_country)
x_agent = agent_encoder.fit_transform(x_agent)

rf_classifier_country = RandomForestClassifier().fit(x_country, y_country)
rf_classifier_agent = RandomForestClassifier().fit(x_agent, y_agent)
```

```python
x_country_missing = missing_country.drop(['Country'], axis=1)
x_agent_missing = missing_agent.drop(['Agent'], axis=1)

x_country_missing = country_encoder.fit_transform(x_country_missing)
x_country_missing.shape
y_country_filled = rf_classifier_country.predict(x_country_missing)
# y_country_filled
```
