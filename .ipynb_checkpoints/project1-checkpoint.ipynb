```python
import pandas as pd

trainData = pd.read_csv('./train.csv', header=None)
testData = pd.read_csv('./test.csv', header=None)

# Remove first NaN column, it seems to serve no purpose
trainData = trainData.iloc[1:, 1:]
testData = testData.iloc[1:, 1:]
```

```python
columns = ['id', 'Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class', 'Flight Distance', 'Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking', 'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment', 'On-board service', 'Leg room service', 'Baggage handling', 'Check-in service', 'Inflight service', 'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes', 'satisfaction']
trainData.columns = columns
testData.columns = columns

trainData.head()
```

```python
from pandas.api.types import is_numeric_dtype

for col in trainData.columns:
    if str.isnumeric(trainData[col].iloc[0]) and col != "id":
        item = pd.to_numeric(trainData[col])
        print('%s:' % (col))
        print('\t Mean = %.2f' % (item.mean()))
        print('\t Standard deviation = %.2f' % (item.std()))
        print('\t Minimum = %.2f' % (item.min()))
        print('\t Maximum = %.2f' % (item.max()))
        print()
```
